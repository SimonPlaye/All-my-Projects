```python
import os
os.chdir("C:/Users/user/Desktop/Data science Invyo/OPCM")

import glob
import sys
sys.path.append("C:/Users/user/AppData/Local/Programs/Python/Python37-32/Lib/site-packages")
sys.path
import PyPDF2 
import re


hours = [str(x) + "h" for x in range(9,24)]

for file in glob.glob("pdf/*.pdf"):
    print(file)

    fname = file
    fname = re.sub(".pdf",".txt", fname)
    ftext = ""
    try:
                
        pdfFileObj = open(file, 'rb') 
    
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
       
        for page in range(pdfReader.numPages):
            pageObj = pdfReader.getPage(page)  
            text = pageObj.extractText()
        
            text = re.sub("\n", "", text)
        
            ##########
        
            text = text.split(".")
        
            for t in text:
                for h in hours:
                    if " " + h in t:
                        ftext += h + " - " + t + " \n"
        


        with open(fname, 'w') as f:
            f.write(ftext)
        
        pdfFileObj.close()
    except:
        pass
```
