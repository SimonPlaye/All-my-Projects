```python
"""Classe servant à récupérer des données de plusieurs actions pour créer des portefeuilles et 
comparer les performances de ces portefeuilles"""
class ManagePtf:
    
    #Récupère les portefeuilles existants s'ils existent
    def __init__(self):
        self.ptf = dict(open("C:\\Users\\simon\\OneDrive\\Bureau\\Tout\\Programmation\\Python\\Portefeuille\\ptf.txt","r").read())
        self.assets = pd.DataFrame()
        
    #Récupère les cours pour un actif
    def inputAsset(self, thicker, debut, fin):
        df = pdr.get_data_yahoo(thicker, start=debut, end=fin)
        df.reset_index(inplace=True)
        df = df[["Date","Close"]]
        df["Date"] = pd.to_datetime(df["Date"], format = "%y%m%d")
        return(df)
    
    #Enregistre les cours d'un actif dans le dossier correspondant au type de cet actif (devise, actions, indices...)
    def saveAsset(self, df, type_asset, name):
        type_asset += "\\"
        df.to_csv(type_asset+name+".csv")
        
    #Actualise le cours d'un actif : rajoute des valeurs au fichier csv s'il existe, sinon elle en télécharge un
    def refreshAsset(self, name, type_asset, debut, fin):
        type_asset += "\\"
        try:
            df = pd.read_csv(type_asset+name+".csv")
            df = df.drop(columns="Unnamed: 0")
        except FileNotFoundError:
            df = self.inputAsset(name, debut, fin)
            self.saveAsset(df, type_asset, name)
            asset_found = False
        else:
            asset_found = True
        
        if asset_found:
            debut_donnee_save = date.fromisoformat(df["Date"][0])
            fin_donnee_save = date.fromisoformat(df["Date"][df["Date"].shape[0]-1])
            debut_input = date.fromisoformat(debut)
            fin_input = date.fromisoformat(fin)
            
            save = False
            if fin_input>fin_donnee_save:
                print(df)
                delta = timedelta(days=1)
                donnees_manquantes = self.inputAsset(name, fin_donnee_save + delta, fin_input)
                donnees_manquantes.drop(index=0, inplace=True)
                df = df.append(donnees_manquantes, ignore_index = True)
                save = True
            if debut_input < debut_donnee_save:
                delta = timedelta(days=1)
                donnees_manquantes = self.inputAsset(name, debut_input, debut_donnee_save-delta)
                donnees_manquantes.drop(index=0, inplace=True)
                df = donnees_manquantes.append(df, ignore_index = True)
                save = True
            if save:
                self.saveAsset(df, type_asset, name)
        return(df)

    
    #Renvoie un data frame des actifs sur la période indiquée, en colonne on ne garde que le close
    def getAssetsDataframe(self, debut, fin, *assets):
  
        self.assets = pdr.get_data_yahoo(assets, start=debut, end=fin)
        self.assets = self.assets[["Close"]]
        if len(assets) == 1:
            self.assets = self.assets.rename(columns={"Close":assets[0]})
        else:
            self.assets = self.assets["Close"]
        print(self.assets)
    
    #Affiche les graphiques du ptf
    def plotAssets(self):
              
        self.assets.plot(title="Absolute evolution of the assets")
        
        first_row = self.assets.iloc[0]/100
        base100 = self.assets / first_row
        
        base100.plot(title="Evolution en base 100")
        
        
```
